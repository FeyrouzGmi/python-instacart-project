```python
# Import libraries

import pandas as pd
import numpy as np
import os
```

```python
# Importing data
ords_prods_merge = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_merged_update.pkl')) 
```

```python
# Creating a subset 
df = ords_prods_merge[:1000000]
```

```python
# Grouping Data with pandas
df.groupby('product_name')
```

```python
# Aggregating Data with agg()
df.groupby('department_id').agg({'order_number': ['mean']})
```

```python
df.groupby('department_id')['order_number'].mean()
```

```python
# Performing Multiple Aggregations
df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

```python
# Aggregating Data with transform()
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head(15)
```

```python
# Creating loyalty Flag
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# Checking 
ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head()
```

```python
5.# Checking the basic statistics of the products prices for each loyalty category 
print('The mean, min and max of prices grouped by the loyalty_flag column in ords_prods_merge:')
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

New customers tend to go for the highest prices, followed by regular customers and finally loyal customers, who buy the least expensive products. However, the maximum price is 99,999, it maybe going to skewed the distribution.

```python
6.# Create a spending flag for each user based on the average price across all their orders 
# creating “spending_flag” column using transform()
ords_prods_merge['average_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Rounding averages 
ords_prods_merge['average_price'] = ords_prods_merge['average_price'].round(2)
```

```python
ords_prods_merge.head()
```

```python
# Creating a 'spending_flag' 
ords_prods_merge.loc[ords_prods_merge['average_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['average_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merge.head()
```

```python
# Checking counts
ords_prods_merge['spending_flag'].value_counts()
```

```python
7.# Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. 
ords_prods_merge['median_prior_orders'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head()
```

```python
ords_prods_merge.loc[ords_prods_merge['median_prior_orders'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_prior_orders'] > 10) & (ords_prods_merge['median_prior_orders'] <= 20), 'order_frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_prior_orders'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```
