```python
# Define a function
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply the function
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
df['prices'].max()
```

```python
# If statement with the loc() function
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df['price_range_loc'].value_counts(dropna = False)
```

```python
4.# Creating a dataframe with the count of values for the column 'orders_hour_of_day'
counts = df_merged['orders_hour_of_day'].value_counts().rename_axis('hour').reset_index(name='counts')
```

```python
counts
```

```python
# Creating a list of percentiles for the value counts of 'orders_hour_of_day'
orders = list(df_merged['orders_hour_of_day'].value_counts().quantile([0.25, 0.5, 0.75]))
```

```python
orders
```

```python
# Fewest orders list
few = list(counts.loc[counts['counts'] <= orders[0]]['hour'])
```

```python
# Average orders list
avg_ords = list(counts[(counts['counts'] > orders[0]) & (counts['counts'] < orders[2])]['hour'])
```

```python
# Most orders list
Mos_ords = list(counts.loc[counts['counts'] >= orders[2]]['hour'])
```

```python
index_few = df_merged['orders_hour_of_day'].isin(few)
```

```python
index_avg_ords = df_merged['orders_hour_of_day'].isin(avg_ords)
```

```python
index_Mos_ords = df_merged['orders_hour_of_day'].isin(Mos_ords)
```

```python
df_merged.loc[index_few,'busiest_period_of_day'] = 'Fewest orders'
```

```python
df_merged.loc[index_avg_ords,'busiest_period_of_day'] = 'Average orders'
```

```python
df_merged.loc[index_Mos_ords,'busiest_period_of_day'] = 'Most orders'
```
